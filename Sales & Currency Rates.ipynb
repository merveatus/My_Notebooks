# Task 04. What is the total value of sales in EUR?

```python
sales1.date = pd.to_datetime(sales1.date)
sales1.date
```

```python
rates1.date = pd.to_datetime(rates1.date)
rates1.date
```

```python
table = pd.merge(sales1, rates1, how="inner", on = ["currency", "date"])
table.head()
```

```python
table["currency"].value_counts()
```

```python
table["Euro_price"] = table.price / table.rate
table.head()
```

```python
table.Euro_price.sum()
```

# Task 05. Which brand has the highest number of purchases during the period?

```python
table.groupby("brand").sum().sort_values(by="Euro_price", ascending=False)
```

```python
table.brand.value_counts()
```

# Task 06. How many items in the “Jewellery” category have no brand associated with them?

# Task 07. How many brands have between 35 and 55 transactions (inclusive)?

```python
((table.brand.value_counts() > 34) & (table.brand.value_counts() < 56)).sum()
```

# Task 09. Which brand has the highest average transaction value?

```python
table.groupby("brand").Euro_price.mean().sort_values(ascending=False).head(1)
```

# Task 11. What percentage of US sellers' transactions were purchased by US buyers?
